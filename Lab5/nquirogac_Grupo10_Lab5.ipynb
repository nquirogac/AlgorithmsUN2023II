# Write a code that produces the the ** Bell numbers** count the possible partitions of a set.

**Hint**:
. Output the n-th Bell Number - [stackexchange](https://codegolf.stackexchange.com/questions/132379/output-the-n-th-bell-number)

```python
def bell(n):
    if n == 0:
        return 1

    b = 0
    for k in range(n):
        b += binomial_coefficient(n - 1, k) * bell(k)

    return b

def binomial_coefficient(n, k):
    if k == 0 or k == n:
        return 1
    else:
        return binomial_coefficient(n - 1, k - 1) + binomial_coefficient(n - 1, k)
```

```python
for i in range(22):
  print(f"The Bell number for {i} is {bell(i)}")
```

###Write a code that that prduces  list of all partititions of the set of $[n]=\{ 1,2,3,4\cdots , n-1, n\}$ with $0 \le n \le 65$ and also list all the partitions for wich the sum of elements of the partions is the same constant for all.

```python
def particiones(conjunto, particion_actual=[]):
    if not conjunto:
        yield particion_actual
        return
    elemento = conjunto[0]
    for particion in particiones(conjunto[1:], particion_actual + [[elemento]]):
        yield particion
    for i, subset in enumerate(particion_actual):
        copia_particion = particion_actual[:]
        copia_particion[i] = subset + [elemento]
        for particion in particiones(conjunto[1:], copia_particion):
            yield particion

def imprimir_particiones(conjunto):
    for particion in particiones(conjunto):
        print(particion)

n = int(input("Ingrese n: "))
conjunto = list(range(1, n+1))
imprimir_particiones(conjunto)
```
